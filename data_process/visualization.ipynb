```python
%matplotlib inline

import matplotlib
from IPython.display import Markdown, display
import numpy as np
import glob
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import pandas as pd
```

```python
dataset_dir = './dataset'
image_files = glob.glob(dataset_dir + '/**/*.jpg')
flower_dir = glob.glob(dataset_dir + '/*')
num_images = len(image_files)
num_classes = len(flower_dir)
```

```python
print("Total: {} samples".format(num_images))
print("Classes: {}".format(num_classes))
print("Samples per class: {}".format(int(num_images / num_classes)))
```
